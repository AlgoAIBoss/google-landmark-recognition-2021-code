```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math
```

```python
landmarks = pd.read_csv('../../Data/train.csv')
landmarks
```

```python
landmarks.id[0]
```

```python
def show_image(image_id):
    img = mpimg.imread('../../Data/train/{}/{}/{}/{}.jpg'.format(image_id[0],image_id[1],image_id[2],image_id))
    return img
```

```python
plt.imshow(show_image(landmarks.id[2]))
plt.show()
```

```python
def show_grid(landmark_id, number = None):
    landmark = landmarks[landmarks.landmark_id == landmark_id].reset_index(drop=True)
    image_list = landmark.id[:number]
    nrows = math.ceil(len(image_list)/4)
    ncols = 4
    fig = plt.gcf()
    fig.set_size_inches(ncols*4, nrows*4)
    sp = plt.subplot(nrows,ncols,len(image_list))
    sp.axis('Off')
    for image in range(0,len(image_list)):
        sp = plt.subplot(nrows,ncols,image+1)
        sp.axis('Off')
        plt.imshow(show_image(image_list[image]))
    plt.show()
```

```python
show_grid(1)
```
